```python

cars_details=[]
i=0
for link in links:
    baseurl='http://'+link
    r = requests.get(baseurl)
    soup = BeautifulSoup(r.text, 'html.parser')
    keyfacts = soup.find_all('span', class_="sc-font-l cldt-stage-primary-keyfact")
    car={}
    try:
        car['mileage']=keyfacts[0].string
        car['registration_date']=keyfacts[1].string
        car['engine_power']=keyfacts[2].string
    except:
        print('no_keyfacts')
    #details
    details=soup.find_all('a', class_="cldt-stealth-link")
    try:
        car['brand']=details[1].string
    except:
        car['brand']=""
    try:
        car['model']=details[2].string
    except:
        car['model']=""
    try:
        car['registration_year']=details[3].string
    except:
        car['registration_year']=""
    try:
        car['colour']=details[4].string
    except:
        car['colour']=""
    try:
        car['type']=details[5].string
    except:
        car['type']=""
    try:
        car['fuel_type']=details[7].string
    except:
        car['fuel_type']=""
      
    ####other details
    det_2=soup.find_all('div',class_="cldt-categorized-data")
    try:
        car['is_after_accident']=det_2[0].find_all('dd')[4].string
    except:
        car['is_after_accident']='no_data_about_accident'
    try:
        car['engine_size']=det_2[2].find_all('dd')[2].string
    except:
        car['engine_size']='no_data'
    print(i)
    i=i+1
    cars_details.append(car)
```
